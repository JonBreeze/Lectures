**Solving the time-independent Schrödinger Equation for a simple 1D potential well**

We shall find the roots of the following characteristic equation:
$$
\frac{\tan(kL)}{k} = -\frac{1}{\lambda} = -\left( \frac{2m}{\hbar^2} V - k^2 \right)^{-\frac{1}{2}}
$$
which we derived for the potential well
<img src="./well.png" width="30%" height="30%">

We shall use $m$ = mass of an electron, $V$ = 20 eV and $L = 4$ Angstrom.  
<br><br>
First things first - load Pylab and configure graphs

```python
# define parameters to prettify graphs
%pylab inline
matplotlib.rcParams['figure.figsize'] = [9,6]
matplotlib.rcParams['font.size'] = 26
matplotlib.rcParams['lines.linewidth'] = 2
matplotlib.rcParams['axes.linewidth'] = 1.5
matplotlib.rcParams['legend.fancybox'] = True
#matplotlib.rcParams['legend.numpoints'] = 1
matplotlib.rcParams['xtick.major.pad'] = 12
matplotlib.rcParams['ytick.major.pad'] = 12

rcParams['axes.labelsize'] = 32
rcParams['xtick.labelsize'] = 20
rcParams['ytick.labelsize'] = 20
rcParams['legend.fontsize'] = 18
```

Now we shall plot $\tan(kL)$ versus $ \frac{k}{\lambda} = -k\left( \frac{2m}{\hbar^2} V - k^2 \right)^{-\frac{1}{2}}$ and find the roots using `fsolve` and initial guesses.


```python
k = linspace(0,3e10,10000) # define range of k values
me = 9.10956e-31 # mass of electron
hbar = 1.054571726e-34 
V = 20.0 * 1.6e-19 # potential -V in bottom of well
L = 4e-10 # width of well

# plot graphs of tan(kL) and -k/sqrt(2*me/hbar/hbar*V-k**2)
fig = figure()
ax = fig.add_subplot(111)
plot(k*1e-10, tan(k*L), 'b.', label='$\\tan(kL)$')
plot(k*1e-10, -k/sqrt(2*me/hbar/hbar*V-k**2), 'r.', label='$-k\\left(\\frac{2m}{\\hbar}V-k^2\\right)^{-\\frac{1}{2}}$')
ylim([-5,0])
xlabel('Wavenumber $k$, [$\AA^{-1}$]')
legend(loc='best')
tight_layout()
grid(True)
ax.xaxis.set_ticks([pi/8,3*pi/8,5*pi/8])
# find roots by defining three functions, f1=tan(kL), f2=-k/sqrt(2*me/hbar/hbar*V-k**2) and f=f1-f2
def f1(k):
    return tan(k*L)
def f2(k):
    return -k/sqrt(2.0*me/hbar/hbar*V-k**2)
def f(k):
    return f1(k) - f2(k)

# import numerical solver
from scipy.optimize import fsolve 

# find root using guesses from graph, print and plot as green circles
k1 = fsolve(f,0.4e10)
print 'k1=', k1*1e-10,'A^-1'
plot( k1*1e-10, f1(k1), 'go', markersize=30, fillstyle='none', markeredgewidth=2 )

k2 = fsolve(f,1.4e10)
print 'k2=', k2*1e-10, 'A^-1'
plot( k2*1e-10, f1(k2), 'go', markersize=30, fillstyle='none', markeredgewidth=2)

k3 = fsolve(f,2.0e10)
print 'k3=', k3*1e-10, 'A^-1'
plot( k3*1e-10, f1(k3), 'go', markersize=30, fillstyle='none', markeredgewidth=2)

#save the graph
savefig('trans.eps', dpi=60)
```

The next step is to plot the wavefunctions to see if they are bound also smooth and continuous at $z=L$. We will also plot the probability densities $\left|\psi(z)\right|^2$.

```python
# plot wavefunctions
x = linspace(0.0,8,1001)
def wfn(k, x):
    x *= 1e-10
    if x < L:
        return sin(k*x)
    else:
        lam = -k/tan(k*L)
        return exp(-lam*x) *sin(k*L)*exp(lam*L)


vwfn = np.vectorize(wfn)
figure()
plot(x, vwfn(k1,x))
plot(x, vwfn(k2,x))
plot(x, vwfn(k3,x))
plot([4,4],[-1,1],'k--')
xlabel('$z\,[\AA]$')
ylabel('$\psi(z)$')

# plot probability densities
figure()
f = subplot(3,1,1)
plot(x, vwfn(k1,x)**2, 'b')
plot([4,4],[-1,1],'k--')
xlabel('$z\,[\AA]$')
ylabel('$|\psi|^2$')
f.axes.get_xaxis().set_visible(False)
f.axes.get_yaxis().set_ticks([0,1])
ylim([0,1])
f.text(5,0.5,'$\epsilon_1$')
f = subplot(3,1,2)
plot(x, vwfn(k2,x)**2, 'g')
plot([4,4],[-1,1],'k--')
xlabel('$z\,[\AA]$')
ylabel('$|\psi|^2$')
f.axes.get_xaxis().set_visible(False)
f.axes.get_yaxis().set_ticks([0,1])
ylim([0,1])
f.text(5,0.5,'$\epsilon_2$')
f = subplot(3,1,3)
plot(x, vwfn(k3,x)**2, 'r')
plot([4,4],[-1,1],'k--')
xlabel('$z\,[\AA]$')
ylabel('$|\psi|^2$')
ylim([0,1])
f.text(5,0.5,'$\epsilon_3$')
f.axes.get_yaxis().set_ticks([0,1])
```

The probability densities need to be normalized.  For instance, the third eigenfunction $\epsilon_3$ has three places where the probability density is equal to one.  That doesn't make sense.  We need to divide the wavefunctions by $\sqrt{\int_0^\infty \left| \psi(z) \right|^2 dz}$.

I will do this the lazy way (using numerical integration).  Analytical expressions are of course easily obtained.  

How do they look now after being normalized?

```python
# normalise wavefunctions 

from scipy import integrate
# lazy numerical integration to find normalization values
n1 = integrate.quad(lambda x: vwfn(k1,x)**2, 0, 10)[0]
n2 = integrate.quad(lambda x: vwfn(k2,x)**2, 0, 10)[0]
n3 = integrate.quad(lambda x: vwfn(k3,x)**2, 0, 10)[0]

# plot the normalized probability densities
f = subplot(3,1,1)
plot(x, vwfn(k1,x)**2/n1, 'b')
plot([4,4],[-1,1],'k--')
xlabel('$z\,[\AA]$')
ylabel('$|\psi|^2$')
f.axes.get_xaxis().set_visible(False)
f.axes.get_yaxis().set_ticks([0,1])
ylim([0,1])
f.text(5,0.5,'$\epsilon_1$')
f = subplot(3,1,2)
plot(x, vwfn(k2,x)**2/n2, 'g')
plot([4,4],[-1,1],'k--')
xlabel('$z\,[\AA]$')
ylabel('$|\psi|^2$')
f.axes.get_xaxis().set_visible(False)
f.axes.get_yaxis().set_ticks([0,1])
ylim([0,1])
f.text(5,0.5,'$\epsilon_2$')
f = subplot(3,1,3)
plot(x, vwfn(k3,x)**2/n3, 'r')
plot([4,4],[-1,1],'k--')
xlabel('$z\,[\AA]$')
ylabel('$|\psi|^2$')
ylim([0,1])
f.text(5,0.5,'$\epsilon_3$')
f.axes.get_yaxis().set_ticks([0,1])

# print the maximum value of each probability density
print max(vwfn(k1,x)**2/n1)
print max(vwfn(k2,x)**2/n2)
print max(vwfn(k3,x)**2/n3)
```
