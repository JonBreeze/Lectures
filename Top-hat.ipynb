**Wave Equation - Plucked String 'Top Hat'**


Here we will plot the general solution for the 'top hat' plucked string problem, using the general series solution that we derived in the lecture:

$$ \large\boxed{
u(x,t) = \frac{2h}{\pi} \sum_{n=1}^\infty \frac{1}{n} \sin \left( \frac{n\pi x}{L} \right) \left[ \cos \left( \frac{n\pi a}{L} \right) - \cos \left( \frac{n\pi(a+b)}{L} \right) \right] \cos \left( \frac{n\pi vt}{L} \right)
}
$$

First things first, load PyLab and prettify the figures generatred by matplotlib.


```python
%pylab inline
matplotlib.rcParams['figure.figsize'] = [9,6]
matplotlib.rcParams['font.size'] = 26
matplotlib.rcParams['lines.linewidth'] = 2
matplotlib.rcParams['axes.linewidth'] = 1.5
matplotlib.rcParams['legend.fancybox'] = True
#matplotlib.rcParams['legend.numpoints'] = 1
matplotlib.rcParams['xtick.major.pad'] = 12
matplotlib.rcParams['ytick.major.pad'] = 12

rcParams['axes.labelsize'] = 32
rcParams['xtick.labelsize'] = 20
rcParams['ytick.labelsize'] = 20
rcParams['legend.fontsize'] = 18
```

Now let's define a function for $u(x,t)$ with $N$ terms in the series, $a=0.45$ and $b=0.1$ for the 'top-hat' function and a velocity $v=0.02$. Here the distance between the stationary ends of the string is set to $L=1$.


```python
def u(x,t,N):
  # N = number of terms
  L = 1.0     # Length of string when at rest
  h = 1.0     # Height of top-hat
  a = 0.45    # top-hat starts here
  b = 0.1     # width of top-hat
  v = 0.02    # wave velocity
  sum = 0.0
  for n in range(1,N+1):
        sum += (1.0/n)*sin(n*pi*x/L)*(cos(n*pi*a/L)-cos(n*pi*(a+b)/L))*cos(n*pi*v*t/L)
        
  return sum * 2.0*h/pi
```

The 'top-hat' function at time $t=0$ looks like ...

```python
# plot string profile at time t=0
x = linspace(0,1.0,100)    # going to plot u(x,t) so need to generate array for x
figure()    # initiate new figure
plot(x, u(x, 0.0, 1000))    # plot u(x,0) as a function of x
xlabel('$x/L$') # label the figure
ylabel('$u/h$')
```

Now, let's let time vary and watch how the profile of the string evolves.  We can plot the profile of the string $u(x,t)$ as a function of time for say $t=0$ to $t=100$, i.e. $t = [0,200]$ in time steps of $\delta t = 0.5$.  Each figure will be saved as a PNG file with a counter in the filename.  The number of terms in the series is 1000, so that it looks nice. 
<br><br>
This may take some time ...

```python
x = linspace(0,1.0,200)
figure()
Nt = 1000    # number of terms in series to take into account (higher the better)
t = 0.0    # start time 
for i in range(0,200):
    clf()    # clear the plot
    plot(x, u(x, t, Nt), label='$t$ = '+str(t))   # plot u(x,t) and label the time
    legend(loc='best') # put a legend somewhere
    xlim([0.0, 1.0]) # define fixed limits to display
    ylim([-1.1, +1.1])
    xlabel('$x/L$') ; ylabel('Amplitude, $u/h$') # labels
    tight_layout() # fix layout
    savefig('wave-%03d.png' % i ) # save the figure sequentially
    t += 0.5 # add time-step to time t
```

We can see that the initial condition is a *standing wave* that consists of left-moving and right-moving waves.  The initial pfoile splits into these counter propagating waves which then propagate towards the boundaries where they undergo a $\pi$ or 180$^\circ$ phase shift due to reflection. You can also see that that waves interfere constructively (in the middle) and destructively (at the boundaries) as waves should due to the *principle of superposition*.
<br><br>
We can also plot a $x-t$ figure, where $x$ is the usual axes and time $t$ is the other.

```python
fig = figure() # new figure
x = linspace(0,1.0,200) # generate X array 
y = linspace(0,200.0,200) # generate Y array
X,Y = meshgrid(x,y) # creat a mesh grid
imgplot = imshow(u(X,Y,100), cmap='bwr', extent=[0,1,0,200], origin='lower', aspect=0.005) # plot the mesh for u(x,y) as an image
xlabel('$x$ [cm]')
ylabel('Time $t$ [s]')
colorbar() # add colorbar
```
