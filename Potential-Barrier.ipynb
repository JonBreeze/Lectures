# Potential Barrier

We will plot the waves for particle with energy $E$ incident upon a barrier of height $V$.  The reflectivity $r$, is given by $$r=\frac{k-k'}{k+k'}$$ and the transmissivity $t$ by $$t=\frac{2k}{k+k'}.$$  The probability reflection coefficient $R$ is given by $$R=\left| r^2 \right|$$ and the probability transmission coefficient $T$ by $$T= \left| t^2 \right|\frac{k'}{k}.$$

The wavenumber $k$ is given by $$k=\sqrt{\frac{2m}{\hbar}\left( E-V \right)}.$$
<br><br>
In region 1, $V=0$ so $$k_1=\sqrt{\frac{2m}{\hbar}E}.$$
<br><br>
Hence, $$ k_2 = \sqrt{k_1^2 - \frac{2m}{\hbar}V}.$$

```julia
using PyPlot
```

```julia
function wave(x)
    k1 = 1.0
    ϕ = 1.9
    if x < 0.0
        return real(exp(1im*k1*x))
    else
        k2 = sqrt(k1^2 - V + 0im)
        return real(exp(1im*k2*x))
    end
end

function Tran(V)
    k1 = 1.0
    k2 = sqrt(k1^2 - V + 0im)
    
    return abs(2*k1/(k1+k2))^2*k2/k1
end

function Refl(V)
    k1 = 1.0
    k2 = sqrt(k1^2 - V + 0im)
    
    return abs((k1-k2)/(k1+k2))^2
end


fig = figure()
plt.rc("axes", titlesize="small", labelsize=32)
```

```julia
V = -2.0
x = linspace(-20.0,20.0,1000)  ;
@vectorize_1arg Float64 wave
plot(x,wave(x))
plot([0,0],[-1,1],"k--")
xlabel(L"$x$")
ylabel(L"$\psi$")
```
